```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df=sns.load_dataset('tips')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Create a binary target: 1 if tip > 20% of total_bill, else 0
df['tip_percent'] = df['tip'] / df['total_bill']
df['high_tip'] = (df['tip_percent'] > 0.2).astype(int)
```

```python
#independance and dependance feature
X = df.drop(['tip', 'tip_percent', 'high_tip'], axis=1)
y = df['high_tip']
```

```python
print("Before dummies - X shape:", X.shape)
print("y shape:", y.shape)
```

```python
# Encode categoricals
X = pd.get_dummies(X, drop_first=True)
print("After dummies - X shape:", X.shape)
```

```python
#train test split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.33,random_state=42)
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
#train logistic regression
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
#Make Predictions & Evaluate
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
coeffs = pd.Series(model.coef_[0], index=X.columns)
print(coeffs.sort_values(ascending=False))
```
